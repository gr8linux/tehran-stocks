## Find a specefic stock


```python
stock_by_symbol = Stocks.query.filter_by(name="كگل").first()  # find by symbol(نماد)
stock_by_title = Stocks.query.filter(
    Stocks.title.like("%گل گهر%")
).first()  # Search by title
stock = stock_by_symbol
stock_by_symbol == stock_by_title
stock
```

# List stocks in a group and call them


```python
stock_list_felez = Stocks.query.filter_by(
    group_code=34
).all()  # find all Stocks in Khodro
stock_list_felez_and_khodro = Stocks.query.filter(
    Stocks.group_code.in_([13, 34])
).all()  # all stocks in khodro and felezat
```

# Advance query


```python
db.session.query(Stocks.group_code, Stocks.group_name).group_by(Stocks.group_code).all()


db.session.execute(
    "select group_code , group_name from stocks group by group_name limit 4"
).fetchall()
```
